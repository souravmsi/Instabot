## Important Libraries

```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time
```

## Starting the Chrome Driver Session

```python
driver = webdriver.Chrome(executable_path = 'c:Downloads/chromedriver.exe')
driver.get('https://www.instagram.com/')
driver.maximize_window()
```

## Entering the Login Credentials

```python
wait = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CLASS_NAME,'f0n8F ')))
username = driver.find_element_by_name('username')
password = driver.find_element_by_name('password')
username.send_keys('SAMPLE_USERNAME')
password.send_keys('SAMPLE_PASSWORD')
password.submit()
```

## Linking First 30 posts of Dilsefoodie

## Function which will return the list of followers of "foodtalkindia" that user is following

```python
def get_followers():
    count=0
    username = []
    while True:
        driver.execute_script('arguments[0].scrollIntoView(0, 100);',driver.find_element_by_xpath('//ul[contains(@class, "jSC57")]'))
        wait = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CLASS_NAME,"wo9IH")))
        elements = driver.find_elements_by_class_name('wo9IH')
        for i in range(count,len(elements)):
            name = elements[i].find_element_by_xpath('.//a[contains(@class,"notranslate")]').get_attribute('innerHTML')
            status = elements[i].find_element_by_xpath('./div/div[2]/button').get_attribute('innerHTML')
            if status == 'Following':
                username.append(name)
                count += 1
            else:
                break
        if status == 'Follow':
            driver.find_element_by_xpath('//div[contains(@class,"WaOAr")]/button').click()
            return username
```

## Function which will return the list of followers of the user

```python
def get_user_followers():
    l=[]
    time.sleep(3)
    followers = driver.find_element_by_xpath('/html/body/div/section/main/div/header/section/ul/li[2]/a')
    followers.click()
    num = int(float(driver.find_elements_by_class_name('g47SY')[1].get_attribute('innerHTML')))
    while True:
        elements = driver.find_elements_by_class_name('wo9IH')
        wait=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//ul[contains(@class, "jSC57")]')))
        if len(elements)<num:
            driver.execute_script('arguments[0].scrollIntoView(0, 100);', driver.find_element_by_xpath('//ul[contains(@class, "jSC57")]'))
            time.sleep(1)
        else:
            break
    for i in elements:
        l.append(i.find_element_by_xpath('.//a[contains(@class,"notranslate")]').get_attribute('innerHTML'))
    driver.find_element_by_xpath('//div[contains(@class,"WaOAr")]/button').click()
    return l
```

```python
try:
    followed_by = driver.find_element_by_xpath('/html/body/div/section/main/div/header/section/div[2]/a/span')
    followed_by.click()
    wait = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CLASS_NAME,"SfPzb")))
    wait.click()
    username_list = get_followers()
    driver.find_element_by_xpath('//div[contains(@class,"_47KiJ")]/div[5]').click()
    time.sleep(1)
    driver.find_element_by_xpath('//div[contains(@class,"_01UL2")]/a').click()
    time.sleep(1)
    user_follower_list = get_user_followers()
    li = set(username_list)-set(user_follower_list)
    for i in li:
        print(i)
except NoSuchElementException:
    print('No such followers to whom, you follow')
```
